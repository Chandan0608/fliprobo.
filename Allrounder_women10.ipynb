```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
source=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder')
soup=bs(source.text,'html.parser')
list=[]
n=soup.find('div',class_='rankings-block__banner--name-large').text
t=soup.find('div',class_='rankings-block__banner--nationality').get_text(strip=True)
r=soup.find('div',class_='rankings-block__banner--rating').text
data={'Name':n,'Team':t,'Rating':r}
list.append(data)
a=soup.find_all('tr',class_='table-body')
for i in a[0:9]:
    n1=i.find('td',class_='table-body__cell rankings-table__name name').a.text
    t1=i.find('span',class_='table-body__logo-text').text
    r1=i.find('td',class_='table-body__cell rating').text
    data={'Name':n1,'Team':t1,'Rating':r1}
    list.append(data)
df=pd.DataFrame(list)
df
    
```

```python
df.to_csv('allrounder_women10.csv')
```
